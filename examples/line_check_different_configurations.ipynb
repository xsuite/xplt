```python
import numpy as np
import xtrack as xt
```

```python
env = xt.Environment()
```

### Horizontal line with thick elements

```python
env.new("quad", "Quadrupole", length=0.3)
env.new("sext", "Sextupole", length=0.2)
env.new("oct", "Octupole", length=0.1)

l1 = env.new_line(
    components=[
        env.new("mb1", "Bend", angle=np.pi / 4, length=1, at=5),
        env.place(["quad", "sext", "oct"]),
        env.new("mb2", "Bend", angle=-np.pi / 4, length=1, at=10),
        env.place(["quad", "sext", "oct"]),
        env.new("end", "Marker", at=15),
    ]
)

sv1 = l1.survey()
```

```python
# Plot ZX
sv1.plot()
```

```python
# Plot ZX
sv1.plot(projection="XZ")
```

### Horizontal line with thin elements

```python
env.new("mm", "Multipole", knl=[0, 3], length=0.3)
l2 = env.new_line(
    components=[
        env.new("m1", "Multipole", hxl=np.pi / 4, knl=[np.pi / 4], length=1, at=5),
        env.place("mm", at=7),
        env.new(
            "m2", "Multipole", hxl=-np.pi / 4, knl=[-np.pi / 4], length=1, at=10
        ),
        env.place("mm", at=12),
        env.new("ee", "Marker", at=15),
    ]
)
sv2 = l2.survey()
```

```python
# ZX plot
sv2.plot()
```

```python
# XZ plot
sv2.plot(projection="XZ")
```

```python
import xplt

xplt.FloorPlot(sv2, line=l2)
```

### Line with thick vertical bends

```python
env.new("bend_v", "Bend", rot_s_rad=np.pi / 2)
l3 = env.new_line(
    components=[
        env.new("mbv1", "bend_v", angle=np.pi / 4, length=1, at=5),
        env.place(["quad", "sext", "oct"]),
        env.new("mbv2", "bend_v", angle=-np.pi / 4, length=1, at=10),
        env.place(["quad", "sext", "oct"]),
        env.new("ev", "Marker", at=15),
    ]
)
sv3 = l3.survey()
```

```python
# Plot ZY
sv3.plot(projection="ZY")
```

```python
l4 = env.new_line(
    components=[
        env.new(
            "mv1", "Multipole", hxl=np.pi / 4, rot_s_rad=np.pi / 2, length=1, at=5
        ),
        env.place("mm", at=7),
        env.new(
            "mv2", "Multipole", hxl=-np.pi / 4, rot_s_rad=np.pi / 2, length=1, at=10
        ),
        env.place("mm", at=12),
        env.new("eev", "Marker", at=15),
    ]
)
sv4 = l4.survey()
```

```python
# Plot ZY
sv4.plot(projection="ZY")
```
